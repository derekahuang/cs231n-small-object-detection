## stats about train dataset

```
with open("datasets/xview/annotations_8020/train.json") as fin:
  data = json.load(fin)
```

```
def plot_stats(areas):
  print('min: ', np.min(areas))
  print('mean: ', np.mean(areas))
  print('median: ', np.median(areas))
  print('max: ', np.max(areas))
  lareas = np.log10(areas)
  plt.hist(lareas, 50)
  plt.xlabel('log(stat)')
  plt.ylabel('count')
  plt.yscale('symlog')
  plt.grid(True)
  plt.show()
```

```
areas = [ann['area'] for ann in data['annotations']]
plot_stats(areas)
```

```
widths = [ann['bbox'][2] for ann in data['annotations']]
plot_stats(widths)
```

```
heights = [ann['bbox'][3] for ann in data['annotations']]
plot_stats(heights)
```

```
anns_per_im = {}
for ann in data['annotations']:
  if ann['image_id'] not in anns_per_im:
    anns_per_im[ann['image_id']] = 0
  anns_per_im[ann['image_id']] += 1

n_anns_per_im = np.array(list(anns_per_im.values()))
plot_stats(n_anns_per_im)
```

```
def get_preds(dirname):
  preds = os.path.join(rdir, dirname, 'inference', 'xview_val_8020', 'predictions.pth')
  x = torch.load(preds)
  return x

preds = get_preds('1-8020-small-anchors')
preds_f = get_preds('1-8020-small-anchors-rpnal-roial-b25-c2b5')
```

```
imgs = {img['id']:img for img in data['images']}
anns = {}
for ann in data['annotations']:
  if ann['image_id'] not in anns:
    anns[ann['image_id']] = []
  anns[ann['image_id']].append(ann)
ids = list(map(lambda t: t[0], sorted(list(map(lambda d: (d['id'], d), data['images'])))))
iimap = {k:v for k,v in enumerate(ids)}
```

```
img_fname = imgs[img_id]['file_name']
print(imgs[img_id])
full_fname = os.path.join('datasets/xview/val_8020', img_fname)
img = Image.open(full_fname)
np_img = np.array(img)
cv_img = np.array(img.convert('RGB'))
cv_img = cv_img[:,:,::-1].copy()
plt.imshow(img)
```

```
img_anns = anns[img_id]
gt_boxes = []
gt_labels = []
for ann in img_anns:
  bbox = ann['bbox']
  bb = [bbox[0], bbox[1], bbox[0]+bbox[2], bbox[1]+bbox[3]]
  gt_boxes.append(bb)
  gt_labels.append(ann['category_id'])

boxes = torch.tensor(gt_boxes)
labels = torch.tensor(gt_labels)

cv_img_cp = cv_img.copy()
for box, label in zip(boxes, labels.int().tolist()):
    box = box.to(torch.int64)
    top_left, bottom_right = box[:2].tolist(), box[2:].tolist()
    cv_img_cp = cv2.rectangle(
        cv_img_cp, tuple(top_left), tuple(bottom_right), tuple(cat_to_color[label]), 1
    )
    t_label = catid_to_name[label]
    t_size = cv2.getTextSize(t_label, 0, 1./3, 1)[0]
    cv_img_cp = cv2.putText(
        cv_img_cp, t_label, (top_left[0], top_left[1]-2), 0, 1./3, [255,255,255]
    )
cv2_imshow(cv_img_cp)
```

```
pr_boxes = preds_f[img_idx]
print(pr_boxes.get_field('scores')[:10])
pr_scores = pr_boxes.get_field('scores') >= 0.3
boxes = pr_boxes.bbox[pr_scores]
pr_labels = pr_boxes.get_field('labels')
labels = pr_labels[pr_scores]

colors = labels[:,None].long() * torch.tensor([2 ** 25 - 1, 2 ** 15 - 1, 2 ** 21 - 1])
colors = (colors % 255).numpy().astype("uint8").tolist()
cv_img_cp = cv_img.copy()
for box, color, label in zip(boxes, colors, labels.int().tolist()):
    box = box.to(torch.int64)
    top_left, bottom_right = box[:2].tolist(), box[2:].tolist()
    catid = contig_to_catid[label]
    cv_img_cp = cv2.rectangle(
        cv_img_cp, tuple(top_left), tuple(bottom_right), tuple(cat_to_color[catid]), 1
    )
    t_label = catid_to_name[catid]
    t_size = cv2.getTextSize(t_label, 0, 1./3, 1)[0]
    cv_img_cp = cv2.putText(
        cv_img_cp, t_label, (top_left[0], top_left[1]-2), 0, 1./3, [255,255,255]
    )
cv2_imshow(cv_img_cp)
```
